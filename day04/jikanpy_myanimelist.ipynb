## Loading jikanpy

```python
from jikanpy import Jikan
import numpy as np
import pandas as pd
import time
```

```python
jikan = Jikan()
```

### Naruto

```python
pages = []
for x in range(1,22):
    pages.append(jikan.anime(id=20, extension='reviews', page=x))
    time.sleep(4)
```

```python
len(pages)
```

```python
df_naruto = pd.DataFrame.from_dict(pages[0]['reviews']) 

for page in pages[1:]: 
  df = pd.DataFrame.from_dict(page['reviews'])
  df_naruto = df_naruto.append(df)

df_naruto
```
